# "Data Preprocessing: Cleaning and Imputing Missing Values"

```python
egy_eco=egy_eco.drop(0)
egy_eco
```

```python
egy_eco['Date'] = egy_eco['Date'].str.replace('M', '', regex=False)
egy_eco['date'] = pd.to_datetime(egy_eco['Date'], format='%Y%m')
egy_eco = egy_eco.drop(columns=['Date'])
print(egy_eco)
```

```python
print(egy_eco.isnull().sum())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline


egy_eco_filled = egy_eco.copy()
if 'date' in egy_eco_filled.columns:
    egy_eco_filled['year'] = egy_eco_filled['date'].dt.year
    egy_eco_filled['month'] = egy_eco_filled['date'].dt.month
    egy_eco_filled['day'] = egy_eco_filled['date'].dt.day
    egy_eco_filled.drop('date', axis=1, inplace=True)

columns_with_missing = egy_eco_filled.columns[egy_eco_filled.isnull().any()].tolist()


for column in columns_with_missing:
    other_columns = [col for col in egy_eco_filled.columns if col != column]
    egy_eco_not_null = egy_eco_filled[egy_eco_filled[column].notnull()]
    egy_eco_null = egy_eco_filled[egy_eco_filled[column].isnull()]
    
    if not egy_eco_not_null.empty:
        model = LinearRegression()
        imputer = SimpleImputer(strategy='mean')
        pipeline = Pipeline([
            ('imputer', imputer),
            ('model', model)
        ])
        
        pipeline.fit(egy_eco_not_null[other_columns], egy_eco_not_null[column])
        predicted_values = pipeline.predict(egy_eco_null[other_columns])
        egy_eco_filled.loc[egy_eco_filled[column].isnull(), column] = predicted_values
```

```python
egy_eco_filled['date'] = pd.to_datetime(egy_eco_filled[['year', 'month', 'day']])
egy_eco_filled.drop(['year', 'month', 'day'], axis=1, inplace=True)
print(egy_eco_filled.head())
```

# "Understanding Data through Descriptive Statistics"

```python
def get_min_max(egy_eco_filled):
    """This function returns the minimum and maximum values for all numeric columns in the data."""
    min_values = {}
    max_values = {}
    numeric_columns = egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        min_values[column] = egy_eco_filled[column].min()
        max_values[column] = egy_eco_filled[column].max()
    return min_values, max_values
    
min_values, max_values = get_min_max(egy_eco_filled)
for column in min_values:
    print(f"{column}: \n Min {min_values[column]}\n Max: {max_values[column]}\n")
```

```python
def get_mean(egy_eco_filled):
    """This function returns the mean values for all numeric columns in the data."""
    mean_values = {}
    numeric_columns = egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        mean_values[column] = egy_eco_filled[column].mean()
    return mean_values

mean_values = get_mean(egy_eco_filled)
for column, mean in mean_values.items():
    print(f"{column}: \n MEAN  {mean_values[column]}\n")
```

```python
def get_mode(egy_eco_filled):
    """This function returns the mode values for all numeric columns in the data."""
    mode_values = {}
    numeric_columns = egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        mode_values[column] = egy_eco_filled[column].mode()[0]
    return mode_values

mode_values = get_mode(egy_eco_filled)
for column, mode in mode_values.items():
    print(f"{column}: \n MODE  {mode_values[column]}\n")
```

```python
def get_range(egy_eco_filled):
    """This function returns the range values for all numeric columns in the data."""
    range_values = {}
    numeric_columns =egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        range_values[column] = egy_eco_filled[column].max()-egy_eco_filled[column].min()
    return range_values

range_values = get_range(egy_eco_filled)
for column, range in range_values.items():
    print(f"{column}: \n RANGE  {range_values[column]}\n")
```

```python
def get_std(egy_eco_filled):
    """This function returns the standard deviation values for all numeric columns in the data."""
    std_values = {}
    numeric_columns =egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        std_values[column] = egy_eco_filled[column].std()
    return std_values

std_values = get_std(egy_eco_filled)
for column, std in std_values.items():
    print(f"{column}: \nSTANDARD DEVIATION : {std}\n")
```

```python
def get_var(egy_eco_filled):
    """This function returns the varience values for all numeric columns in the data."""
    var_values = {}
    numeric_columns = egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        var_values[column] = egy_eco_filled[column].var()
    return var_values

var_values = get_var(egy_eco_filled)
for column, var in var_values.items():
    print(f"{column}: \n VARIENCE  {var_values[column]}\n")
```

```python
import scipy.stats as stats
def skewness(egy_eco_filled):
    """This function returns the Skewness values for all numeric columns in the data."""
    skew_values={}
    numeric_columns =egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        skew_values[column] = egy_eco_filled[column].skew()
    return skew_values
    
skew_values=skewness(egy_eco_filled)
for column, skew in skew_values.items():
    print(f"{column}: \n SKEWNESS  {skew_values[column]}\n")
```

```python
import scipy as stats
def kurtosis(egy_eco_filled):
    """This function returns the Kurtosis values for all numeric columns in the data."""
    kurt_values={}
    numeric_columns =egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        kurt_values[column] =egy_eco_filled[column].kurtosis()
    return kurt_values
    
kurt_values=kurtosis(egy_eco)
for column, kurt in kurt_values.items():
    print(f"{column}: \n KURTOSIS  {kurt_values[column]}\n")
```

```python
def IQR(egy_eco_filled):
    """This function returns the IQR values for all numeric columns in the data."""
    iqr_values = {}
    q1 = {}
    q3 = {}
    lower_bound = {}
    upper_bound = {}
    
    numeric_columns = egy_eco_filled.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        q1[column] = egy_eco_filled[column].quantile(0.25)
        q3[column] = egy_eco_filled[column].quantile(0.75)
        iqr_values[column] = q3[column] - q1[column]
        lower_bound[column] = q1[column] - 1.5 * iqr_values[column]
        upper_bound[column] = q3[column] + 1.5 * iqr_values[column]
    return iqr_values, lower_bound, upper_bound

iqr, lower_bound, upper_bound = IQR(egy_eco_filled)
for column in iqr:
    print(f"{column}:\n IQR {iqr[column]}\n Lower Bound {lower_bound[column]}\n Upper Bound {upper_bound[column]}\n")
```
